```python
def lemmatization(texts, allowed_postags=["NOUN", "ADJ", "VERB", "ADV"]):
    nlp = spacy.load("en_core_web_sm", disable=["parser", "ner"])
    texts_out = []
    for text in texts:

          doc = nlp(text)
          new_text = []
          for token in doc:
            #if "risk" not in text.lower():
              if token.pos_ in allowed_postags:
                  #if "risk" not in token.lemma_:
                  #if token.lemma_!="risk":
                    new_text.append(token.lemma_)
          final = " ".join(new_text)
          texts_out.append(final)
    return (texts_out)


lemmatized_texts = lemmatization(data)
print (lemmatized_texts[0])
```

```python
def gen_words(texts):
    final = []
    for text in texts:
        new = gensim.utils.simple_preprocess(text, deacc=True)
        final.append(new)
    return (final)

data_words = gen_words(lemmatized_texts)

print (data_words[0][0:20])
```

```python
#BIGRAMS AND TRIGRAMS
bigram_phrases = gensim.models.Phrases(data_words, min_count=5, threshold=100)
trigram_phrases = gensim.models.Phrases(bigram_phrases[data_words], threshold=100)

bigram = gensim.models.phrases.Phraser(bigram_phrases)
trigram = gensim.models.phrases.Phraser(trigram_phrases)

def make_bigrams(texts):
    return([bigram[doc] for doc in texts])

def make_trigrams(texts):
    return ([trigram[bigram[doc]] for doc in texts])

data_bigrams = make_bigrams(data_words)
data_bigrams_trigrams = make_trigrams(data_bigrams)

print (data_bigrams_trigrams[0][0:20])
```

```python
def fetch_news_articles(api_key, query, language='en', page_size=5):
    url = 'https://newsapi.org/v2/everything'
    params = {
        'q': query,
        'language': language,
        'pageSize': page_size,
        'apiKey': api_key
    }
    response = requests.get(url, params=params)
    articles = response.json().get('articles', [])
    return articles
```

```python
# Function to find relevant news articles for each company
def find_relevant_articles_for_companies(documents_with_companies, lda_model, dictionary, api_key):
    relevant_articles_by_company = {}

    for document, company in documents_with_companies:
        # Get the topic distribution for the document
        doc_bow = dictionary.doc2bow(document.split())
        doc_topic_distribution = lda_model.get_document_topics(doc_bow)

        # Get the top topics
        top_topics = sorted(doc_topic_distribution, key=lambda x: x[1], reverse=True)

        # Fetch news articles related to the top topics
        for topic_id, _ in top_topics:
            topic_keywords = [word for word, _ in lda_model.show_topic(topic_id)]
            query = ' '.join(topic_keywords)
            articles = fetch_news_articles(api_key, query)

            # Store relevant articles
            if company not in relevant_articles_by_company:
                relevant_articles_by_company[company] = []
            relevant_articles_by_company[company].extend(articles)

    return relevant_articles_by_company
```
